Automatic File Sorter 

```python
import os, shutil
```

```python
path = r"/Users/ivandelapaz/Desktop/Data Analyst youtube course /"
```

```python
os.listdir(path)
```

```python
file_name = os.listdir(path)
```

```python
folder_names = ['csv_file','xlsx_file','sql_file'] 

for loop in range(0,3):
    if not os.path.exists(path + folder_names[loop]):
        print(path + folder_names[loop])
        os.makedirs(path + folder_names[loop])
```

```python
for file in file_name:
    if ".csv" in file and not os.path.exists(path + "csv_file/" + file):
        shutil.move(path + file, path + "csv_file/" + file )
    elif ".xlsx" in file and not os.path.exists(path + "xlsx_file/" + file):
        shutil.move(path + file, path + "xlsx_file/" + file )
    elif ".sql" in file and not os.path.exists(path + "sql_file/" + file):
        shutil.move(path + file, path + "sql_file/" + file )
    else:
        print("no folder made for file")
```
